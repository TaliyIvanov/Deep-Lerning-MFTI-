# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# YOUR CODE
```

```python
cat_data = data[cat_cols]
num_data = data[num_cols]
dummy_feats = pd.get_dummies(cat_data, drop_first=True) # я не буду пользоваться LabelEncoder, OneHotEncoder
X = pd.concat([num_data, dummy_feats], axis=1)
y = data[target_col]
```

```python
(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, 
                                                    train_size=0.8)
```

```python
X_train.shape, X_test.shape
```

```python
cls = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)

params = {'logisticregression__penalty': ['l1','l2','elasticnet'],
          'logisticregression__C': np.linspace(0.001, 100, num=6),
          'logisticregression__max_iter': [800,1000],
          'logisticregression__solver' : ['saga'],
          'logisticregression__l1_ratio' : np.linspace(0.1,1,num=5)

}

clsCV = GridSearchCV(estimator=cls, 
                     param_grid=params,
                     cv=5, 
                     scoring='roc_auc',
                     n_jobs=-1, 
                     verbose=0)
```

```python
clsCV.fit(X_train, y_train)
```

```python
clsCV.best_params_
```

```python
clsCV.best_score_
```

```python
y_pred = clsCV.predict_proba(X_test)

roc_auc_score(y_test, y_pred[:,1])
```

```python
from catboost import CatBoostClassifier, Pool, metrics, cv, MetricVisualizer
```

```python
categorical_features_indices
```

```python
datapool = Pool(data=data.drop(target_col, axis='columns'), label=data[target_col], cat_features=categorical_features_indices)
```

```python
%%time 

fix_params = {"iterations": 100,
              "loss_function": "Logloss",
              'eval_metric': 'AUC:hints=skip_train~false',
              "verbose": 0,
#               'task_type':'GPU'
             }

params = {"depth": np.arange(2, 10),
          'learning_rate': np.linspace(0.03, 0.8, 8),
          'l2_leaf_reg': np.linspace(3, 4, 6),
          'min_data_in_leaf': np.arange(1, 10, 10), 
          'grow_policy': ['SymmetricTree', 'Depthwise', 'Lossguide']
          }


cat = CatBoostClassifier(**fix_params)
# scores = cat.grid_search(params, datapool, cv=3)
scores = cat.randomized_search(params, datapool, cv=4)
```

```python
scores['params']
```

```python
fix_params['iterations'] = 300
best_cat = CatBoostClassifier(**fix_params, **scores['params'])
best_cat.fit(datapool)
roc_auc_score(data[target_col], best_cat.predict_proba(datapool)[:,1])
```
